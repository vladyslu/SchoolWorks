```python
#Vladyslav Luchenko 08/11/2021
#Activity3
import json


with open('laureate.json') as json_file:
    laureateData = json.loads(json_file.read())

#print("Type:", type(laureateData))


#for item in laureateData:
    #for laureate in laureateData[item]:
        #print(laureate)
countries = {}

#check countries
for item in laureateData:
    for laureate in laureateData[item]:
        countryBorn = laureate.get("bornCountry")
        countryDied = laureate.get("diedCountry")
        if countryBorn == countryDied:
            if countryBorn in countries:
                countries[countryBorn] = countries[countryBorn] + 1
            else:
                countries[countryBorn] = 1
        else:
            if countryBorn in countries:
                countries[countryBorn] = countries[countryBorn] + 1
            else:
                countries[countryBorn] = 1
            if countryDied in countries and countryDied is not None:
                countries[countryDied] = countries[countryDied] + 1
            else:
                countries[countryDied] = 1
          
print("")
Keymax = max(countries, key=countries.get)
Keymin = min(countries, key=countries.get)
#Part1. Identify the country with the most laureates.
print("The country with the most laureates is " + Keymax + " : " + str(countries[Keymax]))
#Part2. Identify the country with the fewest laureates.
print("The country with the fewest laureates is " + Keymin + " : " + str(countries[Keymin]))     










#check cities
cities = {}
for item in laureateData:
    for laureate in laureateData[item]:
        cityBorn = laureate.get("bornCity")
        cityDied = laureate.get("diedCity")
        #print("Born " + str(cityBorn) + "Died " + str(cityDied))
        if cityBorn == countryDied:
            if cityBorn in cities:
                cities[cityBorn] = cities[cityBorn] + 1
            else:
                cities[cityBorn] = 1
        else:
            if cityBorn in cities:
                cities[cityBorn] = cities[cityBorn] + 1
            else:
                cities[cityBorn] = 1
            if cityDied in cities and cityDied is not None:
                
                cities[cityDied] = cities[cityDied] + 1
            else:
                cities[cityDied] = 1
          
print("")
KeymaxCity = max(cities, key=cities.get)
KeyminCity = min(cities, key=cities.get)
#Part3. Identify the city with the most laureates.
print("The city with the most laureates is " + KeymaxCity + " : " + str(cities[KeymaxCity]))
#Part4. Identify the city with the fewest laureates.
print("The city with the fewest laureates is " + KeyminCity + " : " + str(cities[KeyminCity]))        



#Part5. the oldest laureate to ever receive a prize (year of prize - year born)
print("")
prizeAges= {}
amountOfPrizes = {}
for item in laureateData:
    for laureate in laureateData[item]:
        for prize in laureate["prizes"]:
            prizeYear = prize.get("year")
            bornYear = laureate.get("born")
            name = str(laureate.get("firstname")) + " " + str(laureate.get("surname"))
            #motivation = prize.get("motivation")
            if bornYear is not None and prizeYear is not None and name is not None:
                #print(prizeYear + " " + bornYear[0:4] + " " + name)
                if bornYear[0:4] != "0000":
                    winnerAge = int(prizeYear) - int(bornYear[0:4])
                    #print(winnerAge)
                    
                    #counting amount of prizes on a person.
                    if name in prizeAges.keys():
                        if winnerAge > prizeAges[name]:
                            prizeAges[name] = winnerAge
                            amountOfPrizes[name] = amountOfPrizes[name] + 1
                    else:
                        prizeAges[name] = winnerAge
                        amountOfPrizes[name] = 1
                        
oldestWinner = max(prizeAges, key=prizeAges.get)
print("The oldest laureate to ever receive a prize is " + oldestWinner + " : " + str(prizeAges[oldestWinner]))



#Part6. How many laureates died in Germany but were not born in Germany?
germanyCounter = 0
#check countries
for item in laureateData:
    for laureate in laureateData[item]:
        countryBorn = laureate.get("bornCountry")
        countryDied = laureate.get("diedCountry")
        if countryBorn != "Germany" and countryDied == "Germany":
            germanyCounter = germanyCounter + 1
            

print("")
print(str(germanyCounter) + " laureates died in Germany but were not born in Germany.")




#Part7. How many laureates died in the USA but were not born in the USA??
USADiedCounter = 0
#check countries
for item in laureateData:
    for laureate in laureateData[item]:
        countryBorn = laureate.get("bornCountry")
        countryDied = laureate.get("diedCountry")
        if countryBorn != "USA" and countryDied == "USA":
            USADiedCounter = USADiedCounter + 1
            

print("")
print(str(USADiedCounter) + " laureates died in the USA but were not born in the USA.")


#Part8. HHow many laureates died in the USA and were born in the USA?
USACounter = 0
#check countries
for item in laureateData:
    for laureate in laureateData[item]:
        countryBorn = laureate.get("bornCountry")
        countryDied = laureate.get("diedCountry")
        if countryBorn == "USA" and countryDied == "USA":
            USACounter = USACounter + 1
            

print("")
print(str(USACounter) + " laureates died in the USA and were born in the USA.")

#Part9. How many laureates were born in Chicago?
ChicagoCounter = 0
for item in laureateData:
    for laureate in laureateData[item]:
        bornCity = laureate.get("bornCity")
        #print(bornCity)
        if bornCity == "Chicago, IL":
            ChicagoCounter = ChicagoCounter + 1
print("")
print(str(ChicagoCounter) + " laureates were born in Chicago.")
            
    

#Part10. Have any laureates won more than one prize?


#I counted amount of prizes on a person in the Part 5 section. 
amountOfMultiWinners = 0;

for prizes in amountOfPrizes.keys():
    if amountOfPrizes[prizes] > 1:
        amountOfMultiWinners = amountOfMultiWinners + 1;
print("")
print(str(amountOfMultiWinners) + " laureates won more than one prize.")
        
    
    
    
#print("")
#print(laureateData)
```

```python
#Vladyslav Luchenko 08/11/2021
#Activity1
import json


with open('restaurant.json') as json_file:
    restaurantData = json.loads(json_file.read())
        
    
#Part1. Compute the average score for each restaurant.
#Part2. Compute the minimum score for each restaurant.
#Part3. Compute the maximum score for each restaurant.
def average(tempList):
    if len(tempList) != 0:
        number = sum(tempList)/(len(tempList))
    else:
        number = 0
    return number


scores = {}
for item in restaurantData:
    for restaurant in restaurantData[item]:
        restaurantName = restaurant.get("name")
        scores.setdefault(restaurantName, [])
        for grade in restaurant["grades"]:
            score = grade.get("score")
            if score is not None:
                scores[restaurantName].append(score)
                #print(str(name) + " " + str(score))
                
print("Restaurant Scores: ")
for restaurant in scores.keys():
    print(restaurant)
    if len(scores[restaurant]) > 0:
        print("Average: " + str(average(scores[restaurant])))
        print("Maximum: " + str(max(scores[restaurant])))
        print("Minimum: " + str(min(scores[restaurant])))           
    else:
        print("Average: 0")
        print("Maximum: 0")
        print("Minimum: 0")        
                
#Part1. Compute the average score for each type of cuisine in each borough.
#Part2. Compute the minimum score for each type of cuisine in each borough.
#Part3. Compute the maximum score for each type of cuisine in each borough.


boroughScores = {}
for item in restaurantData:
    for borough in restaurantData[item]:
        boroughName = borough.get("borough")
        if boroughName not in boroughScores.keys():
            boroughScores.setdefault(boroughName, [])
        for grade in borough["grades"]:
            score = grade.get("score")
            if score is not None:
                boroughScores[boroughName].append(score)
                #print(str(name) + " " + str(score))
            

print("Borough Scores: ")
for borough in boroughScores.keys():
    print(borough)
    if len(boroughScores[borough]) > 0:
        print("Average: " + str(average(boroughScores[borough])))
        print("Maximum: " + str(max(boroughScores[borough])))
        print("Minimum: " + str(min(boroughScores[borough])))       
    else:
        print("Average: 0")
        print("Maximum: 0")
        print("Minimum: 0") 
    

    
#print("")
#print(restaurantData)
```

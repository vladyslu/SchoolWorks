```python
#Luchenko Vladyslav 08/10/2021

class FileItem:
    name = ""
    date = ""
    def __init__(self, sname, sdate, Path):
        self.name = sname
        self.date = sdate
        self.path = Path
    def display(self):
        print("Name: " + self.name)
        print("Path: " + self.path)
        
class CsvFile(FileItem):
    numberOfFields = 0
    numberOfRows = 0
    numberOfColumns = 0
    size = 0
    def  __init__(self, sname, sdate, Path, snumberOfFields, snumberOfRows, snumberOfColumns, Size): 
        FileItem.__init__(self, sname, sdate, Path)
        self.numberOfFields = snumberOfFields
        self.numberOfRows = snumberOfRows
        self.numberOfColumns = snumberOfColumns
        self.size = Size
        
class JpgItem(FileItem):
    resolution = ""
    aspectRatio = ""
    size = 0
    def  __init__(self, sname, sdate, Path, Resolution, AspectRatio, Size): 
        FileItem.__init__(self, sname, sdate, Path)
        self.resolution = Resolution
        self.aspectRatio = AspectRatio
        self.size = Size
    def display(self):
        print("Resolution: " + self.resolution)
class Mp3File(FileItem):
    length = 0.0
    bitRate = 0
    size = 0
    def  __init__(self, sname, sdate, Path, Length, BitRate, Size): 
        FileItem.__init__(self, sname, sdate, Path)
        self.length = Length
        self.bitRate = BitRate
        self.size = Size
        self.path = Path
class Directory(FileItem):
    path = ""
    storage = []
    def  __init__(self, sname, sdate, Path): 
        FileItem.__init__(self, sname, sdate, Path)
    def store(self, item):
        self.storage.append(item)
        
        
#create and store directories
rootFolder = Directory("C:", "01/01/2001", "C:")
documentsFolder = Directory("Documents", "02/01/2001", "C:/Documents")
rootFolder.store(documentsFolder)
musicFolder = Directory("Music", "02/01/2001", "C:/Documents/Music")
documentsFolder.store(musicFolder)
tablesFolder = Directory("Tables", "03/01/2001", "C:/Documents/Tables")
documentsFolder.store(tablesFolder)
secretsFolder = Directory("Secrets", "03/01/2001", "C:/Documents/Secrets")
documentsFolder.store(secretsFolder)
verySecretFolder = Directory("VerySecret", "03/01/2001", "C:/Documents/Secrets/VerySecret")
secretsFolder.store(verySecretFolder)

#create mp3 files
mirvanaMusic = Mp3File("Mirvana.mp3", "05/05/1995", "C:/Documents/Music", 4.4, 2, 14)
mentallicaMusic = Mp3File("Mentallica.mp3", "25/10/1984", "C:/Documents/Music", 4.1, 3, 18)
KidzWithGumsMusic = Mp3File("KidzWithGums-Tortillazz.mp3", "01/04/2005", "C:/Documents/Music", 1.4, 5, 17)
PolkaMusic = Mp3File("Polka.mp3", "04/08/1755", "C:/Documents/Music", 1.1, 8, 11)
#store mp3 files
musicFolder.store(mirvanaMusic)
musicFolder.store(mentallicaMusic)
musicFolder.store(KidzWithGumsMusic)
musicFolder.store(PolkaMusic)

#create csv files
financesTables = CsvFile("Finances.csv", "05/05/1995", "C:/Documents/", 2, 8, 8, 6)
toDoListTables = CsvFile("toDoList.csv", "25/10/1984", "C:/Documents/Tables", 1, 2, 5, 2)
storageTables = CsvFile("storage.csv", "01/04/2005", "C:/Documents/Tables", 4, 7, 8, 1)
colorsTables = CsvFile("colors.csv", "04/08/1755", "C:/Documents/Tables", 4, 4, 8, 7)
#store csv files
documentsFolder.store(mirvanaMusic)
tablesFolder.store(toDoListTables)
tablesFolder.store(storageTables)
tablesFolder.store(colorsTables)

#create jpg files
personalImage = JpgItem("personal.jpg", "05/05/1995", "C:/Documents/Secrets", "640x360", "16:9", 1)
veryPersonalImage = JpgItem("veryPersonal.jpg", "25/10/1984", "C:/Documents/Secrets/VerySecret", "640x360", "16:9", 4)
doNotOpenImage = JpgItem("doNotOpen.jpg", "01/04/2005", "C:/Documents/Secrets/VerySecret", "800x600", "4:3", 8)
passportsImage = JpgItem("Passports.jpg", "04/08/1755", "C:/Documents", "640x360", "16:9", 2)
#store jpg files
documentsFolder.store(passportsImage)
secretsFolder.store(personalImage)
verySecretFolder.store(veryPersonalImage)
verySecretFolder.store(doNotOpenImage)


rootFolder.display()
mirvanaMusic.display()
financesTables.display()
personalImage.display()



```
